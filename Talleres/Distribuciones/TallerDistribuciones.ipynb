1. Generador bionmial
El generador binomial, extrae una distribuci칩n binomial con los parametros admitidos por esa funci칩n.

Formula binomial

Generator.binomial( n , p , size = none )

```python
# n = numero de veces que se repite el experimentos , numero de intentos.
n = 3
# p = a la probabilidad que tienen de ser satisfactorio
p = [0.1, 0.2, 0.5, 0.6, 0.8,0.9]

plt.figure(figsize=(15, 10))
for i in range (len(p)):
  # se realiza la prueba 100 veces
  binomial = rng.binomial(n,p[i],100)
  plt.subplot(3,2,i+1).set_title('n ='+str(n)+'; p ='+str(p[i]))
  plt.hist(binomial, color='b')
plt.show()
```

1. Generador Laplace
Extraiga muestras de Laplace o distribuci칩n exponencial doble con ubicaci칩n (o media) y escala (decadencia) especificadas.


Generator.laplace(loc=0.0, scale=1.0, size=None)

```python
# l = eL valor del punto m치s alto ( pico) de la distribuci칩n  ,El valor predeterminado es 0.
l = [-0.2, -0.1, 0.0, 0.1, 0.2, 0.1] # localizacion
tama침o = 100
s = [1, 2, 5, 6, 8, 9] # escala 

# lambda =  el decaimiento exponencial. El valor predeterminado es 1. Debe ser no negativo.

plt.figure(figsize=(15, 10))
for i in range (len(s)):
  # se realiza la prueba 100 veces
  laplace = rng.laplace(l[i],s[i],tama침o)
  plt.subplot(3,2,i+1).set_title('l ='+str(l[i])+'; s ='+str(s[i]))
  plt.hist(laplace, color = 'r')
plt.show()
#laplace : Se trata de adaptar a la campana de Gauss, y segun variemos puede cambiar la posicion del eje central

#A medida que se aumenta s (lambda), la base de la distribucion aumenta por lo que a valores menores de este , se tendra un ancho menor 
#, y a valores mas bajos uno mas ancho 
```

Generator.gumbel(loc=0.0, scale=1.0, size=None)

```python
plt.figure(figsize=(15, 10))
for i in range (len(p)):
  # se realiza la prueba 100 veces
  Gumbel = rng.gumbel(l[i],s[i],100)
  plt.subplot(3,2,i+1).set_title('l ='+str(l[i])+'; s ='+str(s[i]))
  plt.hist(Gumbel, color ='c')
plt.show()
#Gumbel En esta grafica la mayoria de valores tienden a estar en las primeras pocisiones, pero cuando se varia aumenta su espacio de tendencia
#La campana tendera a presentare en el  eje negativo 

# a medida que se aumenta s , la distribucion presenta un decaimiento cada vez mas  exponencial , esto conlleva a que aumente la base del a distribuci칩n ,
#desplazando la campana de la distribuci칩n 
```

3.Generador Normal
El generador Normal, extrae una serie de muestras de distribuci칩n normal con los parametros admitidos por esa funci칩n

Generator.normal(loc=0.0, scale=1.0, size=None)

```python
plt.figure(figsize=(15, 10))
l = [-0.2, -0.1, 0.0, 0.1, 0.2, 0.9]
s = [1, 2, 5, 6, 8, 9]

for i in range (len(s)):
  # se realiza la prueba 100 veces
  # l = mu , s = sigma 
  Normal = rng.normal(l[i],s[i],100)
  plt.subplot(3,2,i+1).set_title('l ='+str(l[i])+'; s ='+str(s[i]))
  plt.hist(Normal, color='r', density=True)
plt.show()
#aumento s disminuye el valor de la campana y se ancha la base
#l desplaza el centro 

#mu desplaza el centro de la campana  , para un valor mayor de sigma^2  la base de la distribucion aumenta 
#Cuando ponemos valores como 0.9 y 9 respectivamente se ve mas la tendencia a crear la campana, en comparacion de los otros que aumenta el rango de error
```

* Random Binomial	Cuando se cambia p cambia el eje cenral de la campana de izq. a der.
* Random Laplace	Se trata de adaptar a la campana de Gauss, y segun variemos puede cambiar la posicion del eje central
* Random Gumbel	En esta grafica la mayoria de valores tienden a estar en las primeras pocisiones, pero cuando se varia aumenta su espacio de tendencia
* Random Normal	Cuando ponemos valores como 0.9 y 9 respectivamente se ve mas la tendencia a crear la campana, en comparacion de los otros que aumenta el rango de error
